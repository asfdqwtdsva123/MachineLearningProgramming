```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score



# -----------------------------
# 1) 데이터 준비
# -----------------------------

#데이터 col 명 지정
#sepal = 꽃받침, petal = 꽃잎
cols = ["sepal_length", "sepal_width", "petal_length", "petal_width", "label"]

#csv 파일을 읽어들여 df에 저장 / 헤더가 없기 때문에 cols 리스트를 열 이름으로 지정
#dronpa() : 결측치(NaN, Not a Number)가 있는 행 또는 열을 제거하는 함수
# () : 하나라도 포함된 행 전체 제거
# (axis=1) : 결측치가 포함된 열 전체 제거
# (how='all') 모든 값이 결측치인 행만 제거
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/chap3/data/iris.data", header=None, names=cols).dropna()


X = df.drop(columns=["label"])
y = df["label"]

# 테스트 비율은 20%
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# -----------------------------
# 2) 모델 구성
# -----------------------------


dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(n_estimators=200, random_state=42)
lr = LogisticRegression(max_iter=500)

# -----------------------------
# 3) 모델 학습
# -----------------------------

dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
lr.fit(X_train, y_train)

# -----------------------------
# 4) 모델 평가
# -----------------------------

dt_acc = accuracy_score(y_test, dt.predict(X_test))
rf_acc = accuracy_score(y_test, rf.predict(X_test))
lr_acc = accuracy_score(y_test, lr.predict(X_test))

print("=== Test Accuracy ===")
print(f"Decision Tree : {dt_acc:.4f}")
print(f"Random Forest : {rf_acc:.4f}")
print(f"Logistic Reg. : {lr_acc:.4f}")
```
